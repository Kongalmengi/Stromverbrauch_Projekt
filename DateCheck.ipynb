### 독일 행정구역 정보
- Data/german_regions_flat.csv
- 데이터 전처리 필요 : 컬럼명 변경, 필요 컬럼 추출

```python
# 독일 행정구역 정보.
# 데이터 전처 필요. 데이터 컬럼명 변경 필요. 데이터 분할 필요(도시 행정구역 구분하는 부분.)
df_GR_Flat = pd.read_csv('Data/german_regions_flat.csv', sep=';')
df_GR_Flat.head()
```

```python
df_GR_Flat[df_GR_Flat['1_variable_attribute_label'] == 'Flensburg, kreisfreie Stadt']
```

```python
# kreisfreie Stadt 찾기
df_GR_Flat[df_GR_Flat['1_variable_attribute_label'].str.contains('kreisfreie Stadt')].head()
```

```python
# kreisfreie Stadt 이외의 데이터 찾기. ~는 부정형이므로 이를 활용.
df_GR_Flat[~df_GR_Flat['1_variable_attribute_label'].str.contains('kreisfreie Stadt')].head()
```

```python
# landkreis 찾기
df_GR_Flat[df_GR_Flat['1_variable_attribute_label'].str.contains('Landkreis')].head()
```

```python
# bundesland 찾기
df_GR_Flat[df_GR_Flat['1_variable_attribute_label'].str.contains('Bundesland')].head()
```

```python
df_GR_Flat[df_GR_Flat['1_variable_attribute_code']==1001]
# 1_variable_attribute_code 값은 지역코드임을 알 수 있다. 원래는 01001인데, 앞의 0은 생략됨.
```

```python
# 앞에서 지역코드값의 앞부분 0이 생략된 점을 고려하여 해당 부분을 문자열로 불러와 문제 해결.
# 하지만 이 결과가 추후에 데이터를 다루기 용이한 방법이 될지는 미지수.
df_GR_Flat_1 = pd.read_csv('Data/german_regions_flat.csv', sep=';', dtype={'1_variable_attribute_code':str})
df_GR_Flat_1.head()
```
